#### Importation du graphe depuis le dot file

```python
(G, ) = pydot.graph_from_dot_file("../Tests/test/test.c.011t.cfg.dot")
```

#### Récupérations des subgraphs et de leurs noeuds pour le dotfile test

```python
subgraphs = G.get_subgraph_list()
nodes_foo = subgraphs[0].get_nodes()
nodes_main = subgraphs[1].get_nodes()

print(subgraphs[1].get_name())
print(nodes_foo[0])
print(nodes_main[-1].get_attributes())
```

#### Ajout d'un edge pour le graphe

```python
edge = Edge(nodes_foo[0].get_name()+":s", nodes_main[-1].get_name()+":n")
G.add_edge(edge)
print(edge)
```
